### Datatypes

1.Defination: 

+ Data types are a classification of data which tell the      compiler or interpreter how the programmer intends to use  the data.

+ They determine the type of operations the can be performed
on the data ,the values that the data can take, and the    amount of memory neede to store the data.

2.Importance of Data Types in Programming:

Explaination:

+ Datatypes ensure that data is stored in ana efficiency way.
+ They help in performing correct operations on data.
+  Proper use of data types can prevent errors and bugs in the program

```python
## Integer Example
age = 35
type(age)
```

```python
## Floating point datatype
height = 5.11
print(height)
print(type(height))
```

```python
## String datatype example
name = "Ammu"
print(name)
print(type(name))
```
