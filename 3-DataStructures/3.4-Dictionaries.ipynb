### Dictionaries

### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable(e.g, strings, numbers, or tuples), while values can be of any type.

```python
## Creating Dictionaries
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name":"Ammu", "age":20, "grade":25}
print(student)
print(type(student))
```

```python
## Single key is always used!
student = {"name":"Ammu", "age":20, "name":25}
print(student)
```

```python
student = {"name":"Ammu", "age":20, "grade":"A"}
print(student)
```

```python
## Accesing dictionary element
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name', 'Not Available'))
```

```python
## Modifying dictionary elements
## Dictionaries are mutable, so you can add , update or delete elements
print(student)
```

```python
student = {"name":"Ammu", "age":20, "grade":"A"}
student["age"] = 21  ## Updated value for a key
print(student)
student["address"] = "India"  ## Added a new key and value
print(student)
```

```python
student = {'name': 'Ammu', 'age': 21, 'grade': 'A', 'address': 'India'}
del student['grade']  ## Delete key and value pair
print(student)
```

```python
## Dictionary Methods

Keys = student.keys()  ## Get all the keys
print(Keys)
Values = student.values()  ## Get all the values
print(Values)

items = student.items()  ## Get all the key value pairs
print(items)
```

```python
## Shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "Amulya"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy()  ## Shallow copy
print(student_copy1)
print(student)
```

```python
student["name"] = "AMMU"
print(student_copy1)
print(student)
```

```python
## Iterating over dictionaries
## You can use loops to iterate over dictionaries, keys, values, or items

## Iterate over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for values in student.values():
    print(values)
```

```python
## Iterate over key-value pairs
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries
students = {
    "student1": {"name" : "Pranav" , "age" : 21},
    "student2": {"name" : "Deepak" , "age" : 19}
}
print(students)
```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
student.items()
```

```python
## Iterating over nested dictionaries
for student_id , student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key , value in student_info.items():
        print(f"{key} : {value}")
```

```python
## Dictionary Comprehension
squares = {x : x ** 2 for x in range(5)}
print(squares)
```

```python
squares = {}

for i in range(1, 5):
    squares[i] = i ** 2
print(squares)   
```

```python
## Condition dictionary comprehension
evens = {x : x ** 2 for x in range(10) if x % 2 == 0}
print(evens)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in list

numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
frequency = {}

for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1 = {"a" : 1,"b" : 2}
dict2 = {"c" : 3, "d": 4}
merge_dict = {** dict1 , ** dict2}
print(merge_dict)
```

```python
dict1 = {"a" : 1,"b" : 2}
dict2 = {"c" : 3, "d": 4}
merge_dict = dict1 | dict2
print(merge_dict)
```
