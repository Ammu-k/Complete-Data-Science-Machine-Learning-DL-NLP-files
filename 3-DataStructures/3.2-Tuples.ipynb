### Tuples

### Introduction to Tuples

 Explaination:

Tuples are ordred collections of items that are immutable.They are similar to lists, but their immutability makes them different.

```python
## Creating a tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple = (1, "Ammu", 3.14, True)
print(mixed_tuple)
```

```python
## Immutable nature of tuples:
## Tuples are immutable, meaning their elements cannot be changed once assigned.
lst = [1,2,3,4,5]
print(lst)

lst[1] = "Ammu"
print(lst)
```

```python
numbers[2] = "Ammu"
print(numbers)
```

```python
## Tuple Methods:
print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and unpacking tuples:
## Packing

packed_tuple = 1, "Ammu",3.14
print(packed_tuple)
```

```python
## Unpacking
a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers = (1,2,3,4,5,6)
first, * middle , last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple
## Nested List
lst = [[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
lst = [[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1,2,3)),("a","b","c"),(True,False)

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuple
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item , end = " ")
    print()
```

### Conclusion

Tuples are versatile and useful in many real-world scenarios where an immutable and oredered collection of required. They are commonly used in data structures, function arguments and return values, and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your Python code.
