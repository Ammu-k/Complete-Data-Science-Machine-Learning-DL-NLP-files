#### Importing Modules in Python:Modules and Packages


In Python, modules and packages help organize and reuse code.Here's a comprehensive guide on how to import them.

```python
import math 
math.sqrt(16)
```

```python
from math import sqrt, pi
print(sqrt(4))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1, 2, 3, 4])
```

```python
import pandas as pd
```

```python
from math import *  # * is used for calling all functions in math.
print(sqrt(16))
print(pi)
```

```python
from package.maths import addition
addition(2,3)
```

```python
from package import maths
maths.addition(2,3)
```

```python
from Package.maths import addition, subtraction

print(addition(10, 3))
print(subtraction(10, 3))
```
